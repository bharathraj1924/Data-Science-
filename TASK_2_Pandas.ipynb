Combining DataFrames

```python
#Concatenate DataFrames.
import numpy as np
df1 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   index=["1", "2", "3", "4"],
                   columns=["A", "B", "C", "D"])
df2 = pd.DataFrame(np.random.randint(25, size=(6, 4)),
                   index=["5", "6", "7", "8", "9", "10"],
                   columns=["A", "B", "C", "D"])
pd.concat([df1,df2],axis=0)
```

```python
#Merge DataFrames.
pd.merge(df2, df1, on =['A','B','C','D'], how = "outer")
```

Time Series Analysis:


```python
#Convert argument to datetime.
T = pd.DataFrame({'year': [2015, 2016],
                   'month': [2, 3],
                   'day': [4, 5]})
pd.to_datetime(T)
```

```python
#Resample time-series data.
d = {'price': [10, 11, 9, 13, 14, 18, 17, 19],
     'volume': [50, 60, 40, 100, 50, 100, 40, 50]}
df = pd.DataFrame(d)
df['week_starting'] = pd.date_range('01/01/2018',
                                    periods=8,
                                    freq='W')
df.resample('M', on='week_starting').mean()
```

```python
df
```

```python
# Shift index by desired number of periods.
df.shift(periods=2)
```

Statistical Functions:


```python
#Finding Mean
print('Meaan : ',vk['Score'].mean())
#Finding Median
print('Median : ',vk['Score'].median())
#Finding Standard Deviation
print( 'Standard Deviation ',vk['Score'].std())
```

```python
#Compute pairwise correlation.
l= pd.DataFrame([(1, 1), (2, np.nan), (np.nan, 3), (4, 4)],
                  columns=['dogs', 'cats'])
l.corr(min_periods=3)
```

Data Cleaning:


```python
#Rename columns.
l.rename(columns={"dogs" : 'DOGS' , 'cats' : 'CATS' })
```

```python
#Drop specified labels from rows or columns.
l.drop('cats',axis=1)
```
